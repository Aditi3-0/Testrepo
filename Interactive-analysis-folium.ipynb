```python
# Task 2: Mark the success/failed launches for each site on the map
import folium
from folium.features import DivIcon

# Initialize the map centered on the first site
first_site = spacex_df.iloc[0]
site_map = folium.Map(location=[first_site['Lat'], first_site['Long']], zoom_start=5)

# Loop through each launch record
for _, row in spacex_df.iterrows():
    coordinate = [row['Lat'], row['Long']]
    site_name = row['Launch Site']
    outcome = 'Success' if row['class'] == 1 else 'Failure'
    
    # Choose color based on outcome
    color = 'green' if row['class'] == 1 else 'red'
    
    # Add circle with popup
    circle = folium.Circle(
        coordinate,
        radius=1000,
        color=color,
        fill=True
    ).add_child(folium.Popup(f"{site_name}: {outcome}"))
    site_map.add_child(circle)
    
    # Add marker with text label
    marker = folium.map.Marker(
        coordinate,
        icon=DivIcon(
            icon_size=(20,20),
            icon_anchor=(0,0),
            html='<div style="font-size: 12; color:%s;"><b>%s</b></div>' % (color, site_name),
        )
    )
    site_map.add_child(marker)

# Display map in Jupyter Notebook
site_map
```

```python
import math

def haversine(coord1, coord2):
    # Coordinates in [lat, lon]
    lat1, lon1 = coord1
    lat2, lon2 = coord2
    
    # Convert degrees to radians
    lat1, lon1, lat2, lon2 = map(math.radians, [lat1, lon1, lat2, lon2])
    
    # Haversine formula
    dlat = lat2 - lat1
    dlon = lon2 - lon1
    a = math.sin(dlat/2)**2 + math.cos(lat1) * math.cos(lat2) * math.sin(dlon/2)**2
    c = 2 * math.asin(math.sqrt(a))
    
    # Radius of Earth in km
    R = 6371
    return R * c

# Example launch sites
launch_sites = {
  'CCAFS LC-40': [28.562302, -80.577356],
  'CCAFS SLC-40': [28.563197, -80.57682],
  'KSC LC-39A': [28.573255, -80.646895],
  'VAFB SLC-4E': [34.632834, -120.610745]
}

# Example proximities
proximities = {
    'Orlando Airport': [28.4312, -81.3081],
    'Los Angeles Airport': [33.9416, -118.4085],
    'Houston': [29.7604, -95.3698]
}

# Calculate distances
for site_name, site_coord in launch_sites.items():
    print(f"\nDistances from {site_name}:")
    for prox_name, prox_coord in proximities.items():
        distance_km = haversine(site_coord, prox_coord)
        print(f"  → {prox_name}: {distance_km:.2f} km")
```

```python
from math import sin, cos, sqrt, atan2, radians

def calculate_distance(lat1, lon1, lat2, lon2):
    # approximate radius of earth in km
    R = 6373.0

    lat1 = radians(lat1)
    lon1 = radians(lon1)
    lat2 = radians(lat2)
    lon2 = radians(lon2)

    dlon = lon2 - lon1
    dlat = lat2 - lat1

    a = sin(dlat / 2)**2 + cos(lat1) * cos(lat2) * sin(dlon / 2)**2
    c = 2 * atan2(sqrt(a), sqrt(1 - a))

    distance = R * c
    return distance
```

```python
import folium
from folium.features import DivIcon
import math

# Haversine distance function
def calculate_distance(lat1, lon1, lat2, lon2):
    lat1, lon1, lat2, lon2 = map(math.radians, [lat1, lon1, lat2, lon2])
    dlat = lat2 - lat1
    dlon = lon2 - lon1
    a = math.sin(dlat/2)**2 + math.cos(lat1) * math.cos(lat2) * math.sin(dlon/2)**2
    c = 2 * math.asin(math.sqrt(a))
    R = 6371  # Earth radius in km
    return R * c

# Example launch site coordinate
launch_site_lat, launch_site_lon = 28.562302, -80.577356   # CCAFS LC-40

# Example closest coastline coordinate
coastline_lat, coastline_lon = 28.56367, -80.57163

# Calculate distance
distance = calculate_distance(launch_site_lat, launch_site_lon, coastline_lat, coastline_lon)

# Initialize map centered on launch site
site_map = folium.Map(location=[launch_site_lat, launch_site_lon], zoom_start=12)

# Add launch site marker
folium.Marker([launch_site_lat, launch_site_lon], popup="Launch Site").add_to(site_map)

# Add coastline marker with distance label
distance_marker = folium.Marker(
    [coastline_lat, coastline_lon],
    icon=DivIcon(
        icon_size=(20,20),
        icon_anchor=(0,0),
        html='<div style="font-size: 12; color:#d35400;"><b>%s</b></div>' % "{:10.2f} KM".format(distance),
    )
)
site_map.add_child(distance_marker)

# Draw line between launch site and coastline
folium.PolyLine([[launch_site_lat, launch_site_lon], [coastline_lat, coastline_lon]],
                color='blue', weight=2.5, opacity=1).add_to(site_map)

# Display in Jupyter Notebook
site_map
```
